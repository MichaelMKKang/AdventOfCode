```python
import math

#i is the index: the start of the first number where value > 9
def split(snail,i):
    num = int(snail[i:i+2])
    left = math.floor(num/2)
    right = math.ceil(num/2)
    return snail[:i] + '[' + str(left) + ',' + str(right) + ']' + snail[i+2:]

# test = '[[[[0,7],4],[15,[0,13]]],[1,1]]'
# round2 = split(test,13)
# print(round2)
# print(split(round2,22))

assert split('[[[[0,7],4],[15,[0,13]]],[1,1]]',13) == '[[[[0,7],4],[[7,8],[0,13]]],[1,1]]', 'check0'
assert split('[[[[0,7],4],[[7,8],[0,13]]],[1,1]]',22) == '[[[[0,7],4],[[7,8],[0,[6,7]]]],[1,1]]', 'check1'
```

```python
def step(line):
    depth = 0
    max_depth = 4
    max_depth_index = 0
    num = ''
    exploded = False
    for i in range(len(line)):
        if depth > max_depth:
            exploded = True
            max_depth = depth
            max_depth_index = i
        if line[i] == '[':
            depth += 1
        if line[i] == ']':
            depth -= 1
    if max_depth >= 5 and exploded:
        #print('max depth is',max_depth,'at index',max_depth_index)
        line = explode(line,max_depth_index)
        return line
    for i in range(len(line)):
        if line[i] in '0123456789':
            num += line[i]
            if int(num) > 9:
                #print(i)
                line = split(line,i-1)
                return line
        elif line[i] not in '0123456789' and num != '':
            num = ''
    return line

#step1: run this
# test = '[[[[0,[4,5]],[0,0]],[[[4,5],[2,6]],[9,5]]],[7,[[[3,7],[4,3]],[[6,3],[8,8]]]]]'
# for i in range(20):
#     z = step(test)
#     print(z)
#     test = z
 
#step2: this is the issue line
# issue = '[[[[4,0],[5,4]],[[7,7],[6,0]]],[[7,[5,5]],[[0,[[5,6],3]],[[6,3],[8,8]]]]]'
# print(issue[47:51])
    
#print(step('[[[[0,[4,5]],[0,0]],[[[4,5],[2,6]],[9,5]]],[7,[[[3,7],[4,3]],[[6,3],[8,8]]]]]'))
#print(step(step('[[[[0,[4,5]],[0,0]],[[[4,5],[2,6]],[9,5]]],[7,[[[3,7],[4,3]],[[6,3],[8,8]]]]]')))
```

```python
def snail_add(left,right):
    line = '[' + left + ',' + right + ']'
    #print(line)
    while 1:
        temp_line = step(line)
        if temp_line == line:
            break
        else:
            line = temp_line[:]
            #print(line)
    return line

assert snail_add('[[[[4,3],4],4],[7,[[8,4],9]]]','[1,1]') == '[[[[0,7],4],[[7,8],[6,0]]],[8,1]]', 'check combo0'
```

```python
def get_magnitude(line):
    depth = 0
    max_depth = 0
    max_depth_index = 0
    exploded = False
    for i in range(len(line)):
        if depth > max_depth:
            exploded = True
            max_depth = depth
            max_depth_index = i
        if line[i] == '[':
            depth += 1
        if line[i] == ']':
            depth -= 1
    #print(max_depth,max_depth_index)
    #print(line[max_depth])
    
    if max_depth == 0:
        return line
    
    comma_index = max_depth_index
    while line[comma_index] != ',':
        comma_index += 1
    #print(comma_index)
        
    end_index = comma_index
    while line[end_index] != ']':
        end_index += 1
    
    left_num = int(line[max_depth_index:comma_index])
    right_num = int(line[comma_index+1:end_index])
    
    #print(line[:max_depth_index-1])
    #print(left_num*3 + right_num*2)
    #print(line[end_index+1:])
    new_line = line[:max_depth_index-1] + str(left_num*3 + right_num*2) + line[end_index+1:]
    #print('new line:         ',new_line)
    return get_magnitude(new_line)

#get_magnitude(line)
assert get_magnitude('[[1,2],[[3,4],5]]') == '143','check example0'
assert get_magnitude('[[[[0,7],4],[[7,8],[6,0]]],[8,1]]') == '1384','check example1'
assert get_magnitude('[[[[1,1],[2,2]],[3,3]],[4,4]]') == '445','check example2'
assert get_magnitude('[[[[3,0],[5,3]],[4,4]],[5,5]]') == '791','check example3'
assert get_magnitude('[[[[5,0],[7,4]],[5,5]],[6,6]]') == '1137','check example4'
assert get_magnitude('[[[[8,7],[7,7]],[[8,6],[7,7]]],[[[0,7],[6,6]],[8,7]]]') == '3488','echeck example5'
```
